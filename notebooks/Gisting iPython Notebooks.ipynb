- see https://gist.github.com/
- Github [Gist](http://en.wikipedia.org/wiki/GitHub#Gist) is a pastebin-style site for saving text snippets 
- its free
- Gists can be public or secret (unguessable hash hidden form search engines)
- files cant be too big


Installation of a Gister
------------------

- We will use the great, simple  **gister** app https://github.com/ConradIrwin/jist to write notebooks (or any other file) to gist.github.com
- this requires installation of Ruby 
- you will of course need a user account at github.com
- this even works for Windows 

Install the Ruby gem

````python
!sudo gem install gist
````

#### iPython Notebooks are just plain-text [JSON](http://en.wikipedia.org/wiki/JSON) files

Lets look at the top 25 lines of this one 

````python
!head -n 25 Gisting\ iPython\ Notebooks.ipynb
````

### Save this Notebook as a Public Gist

````python
!gist  -d "How to Gist an iPython Notebook" Gisting\ iPython\ Notebooks.ipynb
````

This Gist is now visible

- as the raw Gist at https://gist.github.com/a44c48e20a78e704ba1f 
- rendered as HTML at http://nbviewer.ipython.org/gist/cfljam/a44c48e20a78e704ba1f

so you can store and share a link with this rendering URL

You can get NBViewer extensions and a bookmarklet for Chrome and Firefox at http://jiffyclub.github.io/open-in-nbviewer/ 

### Updating the Gist

Just pass the filename and the URL https://gist.github.com/a44c48e20a78e704ba1f to gist with the update (-u) flag

````python
!gist -u https://gist.github.com/a44c48e20a78e704ba1f Gisting\ iPython\ Notebooks.ipynb
````

If you check the Gist URL, https://gist.github.com/cfljam/a44c48e20a78e704ba1f you can see that these changes are reflected in the revision history 

**Gists** are a simple versioned file storage solution tat can be treated like an GitHub repository.
